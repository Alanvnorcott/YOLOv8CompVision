```python
from ultralytics import YOLO
import sys
import cv2
```

```python
model   = YOLO('yolov8n.pt')
```

```python
image = cv2.imread('/content/pictures/car.jpg')
```

```python
results = model('/content/pictures/car.jpg')
```

```python
# Loop through the detections in the results
for result in results:
    # Access the boxes within the result
    for box in result.boxes:
        # Get the bounding box coordinates
        x1, y1, x2, y2 = box.xyxy[0].tolist()  # Convert tensor to list


        # Draw the bounding box on the image
        cv2.rectangle(image, (int(x1), int(y1)), (int(x2), int(y2)), (0, 255, 0), 2)


        # Optional: Add text with class and confidence
        conf = box.conf[0].item()
        class_id = int(box.cls[0].item())
        label = f'{class_id} {conf:.2f}'
        cv2.putText(image, label, (int(x1), int(y1) - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)
```
